```python
import torch
import torch.nn as nn
from torchtext.vocab import GloVe
from torchtext.datasets import AG_NEWS
from torch.utils.data import DataLoader, Dataset
from torch.nn.utils.rnn import pad_sequence
from torchtext.transforms import VocabTransform
from torchtext.vocab import build_vocab_from_iterator
import spacy
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
import seaborn as sns

# Set random seed for reproducibility
SEED = 1234
torch.manual_seed(SEED)
torch.backends.cudnn.deterministic = True

# Data preprocessing
nlp = spacy.load('en_core_web_sm')
tokenizer = lambda x: [token.text for token in nlp.tokenizer(x)]

# Build vocabulary
def yield_tokens(data_iter):
    for _, text in data_iter:
        yield tokenizer(text)

# Download AG News dataset
train_iter, test_iter = AG_NEWS(split=('train', 'test'))

# Build vocabulary
vocab = build_vocab_from_iterator(yield_tokens(train_iter), specials=["<unk>", "<pad>", "<bos>", "<eos>"])
vocab.set_default_index(vocab["<unk>"])

# Set vocab and label transforms
text_transform = VocabTransform(vocab)
label_transform = lambda x: x - 1  # AG_NEWS labels are 1-4, we need 0-3

# Define dataset class
class AGNewsDataset(Dataset):
    def __init__(self, split):
        self.data = list(AG_NEWS(split=split))
        self.text_transform = text_transform
        self.label_transform = label_transform

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        label, text = self.data[idx]
        return torch.tensor(self.text_transform(tokenizer(text))), torch.tensor(self.label_transform(label))

def collate_batch(batch):
    label_list, text_list, lengths = [], [], []
    for (_text, _label) in batch:
        label_list.append(_label)
        processed_text = _text.clone().detach()
        text_list.append(processed_text)
        lengths.append(processed_text.size(0))
    text_list = pad_sequence(text_list, padding_value=vocab["<pad>"])
    return torch.tensor(label_list, dtype=torch.long), text_list, torch.tensor(lengths)

# Data loading
BATCH_SIZE = 64

train_dataset = AGNewsDataset(split='train')
test_dataset = AGNewsDataset(split='test')

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, collate_fn=collate_batch)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, collate_fn=collate_batch)

# LSTM model definition
class LSTM(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_dim, n_layers,
                bidirectional, dropout, pad_idx):

        super().__init__()

        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=pad_idx)

        self.lstm = nn.LSTM(embedding_dim,
                            hidden_dim,
                            num_layers=n_layers,
                            bidirectional=bidirectional,
                            dropout=dropout)

        self.fc = nn.Linear(hidden_dim * 2 if bidirectional else hidden_dim, output_dim)

        self.dropout = nn.Dropout(dropout)

    def forward(self, text, text_lengths):

        embedded = self.dropout(self.embedding(text))

        packed_embedded = nn.utils.rnn.pack_padded_sequence(embedded, text_lengths.cpu(), enforce_sorted=False)

        packed_output, (hidden, cell) = self.lstm(packed_embedded)

        if self.lstm.bidirectional:
            hidden = self.dropout(torch.cat((hidden[-2, :, :], hidden[-1, :, :]), dim=1))
        else:
            hidden = self.dropout(hidden[-1, :, :])

        return self.fc(hidden)

# ACO (Ant Colony Optimization) implementation
class ACO:
    def __init__(self, n_ants, n_iterations, bounds, embedding_dim, bidirectional, data_loaders, device, SEED, alpha=1, beta=1, evaporation_rate=0.5):
        self.n_ants = n_ants
        self.n_iterations = n_iterations
        self.bounds = bounds
        self.embedding_dim = embedding_dim
        self.bidirectional = bidirectional
        self.data_loaders = data_loaders
        self.device = device
        self.SEED = SEED
        self.alpha = alpha
        self.beta = beta
        self.evaporation_rate = evaporation_rate

        # Initialize pheromone matrix and heuristic information
        self.pheromone = np.ones((len(bounds), self.n_ants))
        self.heuristic = np.random.rand(len(bounds), self.n_ants)

    def fitness(self, params):
        # Unpack parameters
        HIDDEN_DIM, N_LAYERS, DROPOUT = int(params[0]), int(params[1]), params[2]

        # Define model
        INPUT_DIM = len(vocab)
        OUTPUT_DIM = 4  # AG_NEWS has 4 classes
        PAD_IDX = vocab["<pad>"]

        model = LSTM(INPUT_DIM, self.embedding_dim, HIDDEN_DIM, OUTPUT_DIM, N_LAYERS,
                     self.bidirectional, DROPOUT, PAD_IDX).to(self.device)

        # Load pre-trained word vectors
        glove_vectors = GloVe(name='6B', dim=self.embedding_dim)
        pretrained_embeddings = glove_vectors.get_vecs_by_tokens(vocab.get_itos())
        model.embedding.weight.data.copy_(pretrained_embeddings)

        optimizer = torch.optim.Adam(model.parameters())
        criterion = nn.CrossEntropyLoss().to(self.device)

        # Training
        model.train()
        epoch_loss = 0
        train_loader = self.data_loaders  # Use training data loader

        for batch in train_loader:
            optimizer.zero_grad()
            labels, text, text_lengths = batch
            labels, text, text_lengths = labels.to(self.device), text.to(self.device), text_lengths.to(self.device)
            predictions = model(text, text_lengths)
            loss = criterion(predictions, labels)
            loss.backward()
            optimizer.step()
            epoch_loss += loss.item()

        return epoch_loss / len(train_loader)

    def optimize(self):
        best_fitness = float('inf')
        best_params = None

        for iteration in range(self.n_iterations):
            ants_solutions = []
            ants_fitness = []

            for i in range(self.n_ants):
                ant_solution = []
                for j in range(len(self.bounds)):
                    probabilities = (self.pheromone[j, :] ** self.alpha) * (self.heuristic[j, :] ** self.beta)
                    probabilities = probabilities / probabilities.sum()
                    selected_index = np.random.choice(range(self.n_ants), p=probabilities)
                    param_value = self.bounds[j][0] + (self.bounds[j][1] - self.bounds[j][0]) * selected_index / (self.n_ants - 1)
                    ant_solution.append(param_value)

                ants_solutions.append(ant_solution)
                fitness = self.fitness(ant_solution)
                ants_fitness.append(fitness)

                if fitness < best_fitness:
                    best_fitness = fitness
                    best_params = ant_solution

            # Update pheromone levels
            for j in range(len(self.bounds)):
                self.pheromone[j, :] = (1 - self.evaporation_rate) * self.pheromone[j, :]
                for i in range(self.n_ants):
                    self.pheromone[j, i] += 1.0 / ants_fitness[i]

            print(f'Iteration {iteration + 1}/{self.n_iterations}, Best Fitness: {best_fitness}')

        return best_params

# Define hyperparameter bounds and embedding dimension
embedding_dim = 50
bounds = [(128, 512), (2, 4), (0.1, 0.5)]  # Ensure N_LAYERS starts from 2

# Initialize ACO and optimize
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
bidirectional = True
aco = ACO(n_ants=10, n_iterations=3, bounds=bounds, embedding_dim=embedding_dim, bidirectional=bidirectional, data_loaders=train_loader, device=device, SEED=SEED)
best_params = aco.optimize()
print(f'Best Parameters: {best_params}')

# Re-train the model with best parameters and evaluate
HIDDEN_DIM, N_LAYERS, DROPOUT = int(best_params[0]), int(best_params[1]), best_params[2]
INPUT_DIM = len(vocab)
OUTPUT_DIM = 4  # AG_NEWS has 4 classes
PAD_IDX = vocab["<pad>"]

model = LSTM(INPUT_DIM, embedding_dim, HIDDEN_DIM, OUTPUT_DIM, N_LAYERS,
             bidirectional, DROPOUT, PAD_IDX).to(device)

# Load pre-trained word vectors
glove_vectors = GloVe(name='6B', dim=embedding_dim)
pretrained_embeddings = glove_vectors.get_vecs_by_tokens(vocab.get_itos())
model.embedding.weight.data.copy_(pretrained_embeddings)

optimizer = torch.optim.Adam(model.parameters())
criterion = nn.CrossEntropyLoss()

# Training and evaluation
NUM_EPOCHS = 3  # Set the number of epochs to 3
train_losses = []
test_losses = []
best_test_loss = float('inf')

for epoch in range(NUM_EPOCHS):

    # Train the model
    model.train()
    epoch_train_loss = 0

    for batch in train_loader:
        optimizer.zero_grad()
        labels, text, text_lengths = batch
        labels, text, text_lengths = labels.to(device), text.to(device), text_lengths.to(device)
        predictions = model(text, text_lengths)
        loss = criterion(predictions, labels)
        loss.backward()
        optimizer.step()
        epoch_train_loss += loss.item()

    train_losses.append(epoch_train_loss / len(train_loader))

    # Evaluate the model
    model.eval()
    epoch_test_loss = 0
    all_labels = []
    all_predictions = []

    with torch.no_grad():
        for batch in test_loader:
            labels, text, text_lengths = batch
            labels, text, text_lengths = labels.to(device), text.to(device), text_lengths.to(device)
            predictions = model(text, text_lengths)
            loss = criterion(predictions, labels)
            epoch_test_loss += loss.item()

            all_labels.extend(labels.cpu().numpy())
            all_predictions.extend(torch.argmax(predictions, dim=1).cpu().numpy())

    test_losses.append(epoch_test_loss / len(test_loader))

    # Print epoch statistics
    print(f'Epoch [{epoch + 1}/{NUM_EPOCHS}], '
          f'Train Loss: {train_losses[-1]:.4f}, '
          f'Test Loss: {test_losses[-1]:.4f}')

    # Save the model if test loss has decreased
    if test_losses[-1] < best_test_loss:
        best_test_loss = test_losses[-1]
        torch.save(model.state_dict(), 'best_model.pt')

# Evaluation metrics
accuracy = accuracy_score(all_labels, all_predictions)
precision, recall, f1, _ = precision_recall_fscore_support(all_labels, all_predictions, average='weighted')
conf_matrix = confusion_matrix(all_labels, all_predictions)

print(f'\nFinal Evaluation Metrics:')
print(f'Test Accuracy: {accuracy:.4f}')
print(f'Test Precision: {precision:.4f}')
print(f'Test Recall: {recall:.4f}')
print(f'Test F1 Score: {f1:.4f}')

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['World', 'Sports', 'Business', 'Sci/Tech'], yticklabels=['World', 'Sports', 'Business', 'Sci/Tech'])
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```
